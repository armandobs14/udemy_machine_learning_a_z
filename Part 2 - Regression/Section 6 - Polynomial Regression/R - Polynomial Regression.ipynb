# Polynomial Regression - R

## Importing the dataset

```r
dataset = read.csv('Polynomial_Regression/Position_Salaries.csv')
dataset = dataset[2:3]
# dataset
```

 ## Fitting Linear Regression to the dataset

```r
linear_model = lm( formula = Salary ~ ., data = dataset)
```

 ## Fitting Polynomial Regression to the dataset

```r
polynomial_model = lm(Salary ~ poly(Level,4),
          data = dataset)
summary(polynomial_model)
```

```r
# install.packages('ggplot2')
library(ggplot2)
x_grid = seq(min(dataset$Level), max(dataset$Level),0.1)
ggplot() +
    geom_point(aes(x = dataset$Level, y = dataset$Salary),
              colour = 'red') +
    geom_line(aes(x = x_grid, y = predict(polynomial_model, newdata = data.frame(Level = x_grid))),
              colour = 'blue') + 
    ggtitle('Truth or Bluff (Polynomial Regression)') +
    xlab('Level') +
    ylab('Salary')
```

## Predicting a new result with Linear and Polynomial Regression

```r
#linear regression prediction
linear_level = data.frame(Level = 6.5)
linear_level$Prediction = predict(linear_model, linear_level)
print(linear_level)
```

```r
#polynomial regression prediction
poly_level = data.frame(Level = 6.5)
poly_level$Predict = predict(polynomial_model, poly_level)
print(poly_level)
```
