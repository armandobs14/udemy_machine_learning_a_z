# Random forest Regression

```r
#install.packages('rpart')
#install.packages('randomForest')
library(rpart)
library(ggplot2)
library(randomForest)
```

```r
dataset = read.csv('Random_Forest_Regression/Position_Salaries.csv')
dataset = dataset[2:3]

# For reproducibility
set.seed(1234)

model = randomForest(x = dataset[1],
                     y = dataset$Salary,
                    ntree = 300)
```

```r
predict(model, data.frame(Level = 6.5))
```

```r
# Visualising SVR results
x_grid = seq(min(dataset$Level), max(dataset$Level),0.01)
ggplot() +
    geom_point( aes(x = dataset$Level, y = dataset$Salary), colour = 'red') +
    geom_line( aes(x = x_grid, y = predict(model, newdata = data.frame(Level = x_grid))), colour = 'blue') +
    ggtitle('Truth or Bluff (Random Forest Regression)') +
    xlab('Label') +
    ylab('Salary')
```
