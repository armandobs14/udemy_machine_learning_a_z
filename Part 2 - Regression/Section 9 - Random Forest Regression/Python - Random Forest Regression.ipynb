# Random forest Regression

```python
# importing libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
# Importing dataset and split into independent and dependent variables
df = pd.read_csv('Random_Forest_Regression/Position_Salaries.csv')
X = df.iloc[:,1:2].values
Y = df.iloc[:,2].values
```

```python
# Creating regressor
model = RandomForestRegressor(n_estimators = 300, random_state = 0)
model.fit(X, Y)
```

```python
model.predict(6.5)
```

```python
# Visualizing Decision Tree (Low Resolution)
plt.scatter(X, Y, color = 'red')
plt.plot(X, model.predict(X), color = 'blue')
plt.title('Truth or bluff (Decision Tree Regression) - Low Resolution')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()


# Visualizing Decision Tree (High Resolution)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(X, Y, color = 'red')
plt.plot(X_grid, model.predict(X_grid), color = 'blue')
plt.title('Truth or bluff (Decision Tree Regression) - High Resolution')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
