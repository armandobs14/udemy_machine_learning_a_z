# Decision Tree Regression

```r
#install.packages('rpart')
library(rpart)
library(ggplot2)
```

```r
dataset = read.csv('Decision_Tree_Regression/Position_Salaries.csv')
dataset = dataset[2:3]
model = rpart(formula = Salary ~ ., dataset, control = rpart.control(minsplit = 1))
```

```r
predict(model, data.frame(Level = 6.5))
```

```r
# Visualising SVR results
x_grid = seq(min(dataset$Level), max(dataset$Level),0.01)
ggplot() +
    geom_point( aes(x = dataset$Level, y = dataset$Salary), colour = 'red') +
    geom_line( aes(x = x_grid, y = predict(model, newdata = data.frame(Level = x_grid))), colour = 'blue') +
    ggtitle('Truth or Bluff (SVR)') +
    xlab('Label') +
    ylab('Salary')
```
